# Gussian Funtions: Gaussian Distribution and Their Plots

#### Content:
    -1D Gaussian (Input: mu and variance)
    -2D Gaussian (Input: mu of x and y, cariance of x and y)

### Inport inbuilt Fuction

```python
%matplotlib inline
import numpy as np
import math
from pylab import *
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

### 1-D Guassian Distribution Function

```python
def D_1Gaussian (mu, var, x):
    
    pdf = 1/math.sqrt(2.*math.pi*var) * math.exp(-0.5*(x-mu)**2 / var)

    return pdf
```

### 1-D Gussian Distribution Plot Displacement

```python
#Adjust mu, var and the range of x

mu = 6
var = 5
x = np.linspace (0,20,100)
y = []

for i in range(len(x)):
    y.append(D_1Gaussian (mu, var, x[i]))
    
#Display the plot
figure()
xlabel('x')
ylabel('y')
plot(x,y,'r')
title('1-D Gaussian')
show()
```

### 2-D Gaussian Distribution Function

```python
def D_2Gaussian (mu_x, mu_y, var_x, var_y, x, y):
    
    pdf = (1/(2.*math.pi*var_x*var_y))*math.exp(-(((x-mu_x)**2)/(2*(var_x**2))+((y-mu_y)**2)/(2*(var_y**2))))
    
    return pdf 
```

### 2-D Guassian Distribution Displacement

```python
#Adjust mu_x, mu_y, var_x, var_y, and the range of variables x and y

mu_x = 8
mu_y = 12
var_x = 6
var_y = 4
x = np.linspace (0,20,500)
y = np.linspace (0,20,500)
a = []
for i in range(len(x)):
    for j in range (len(y)):
        a.append(D_2Gaussian(mu_x, mu_y, var_x, var_y, x[i], y[j]))

z = np.reshape(a,(len(x),len(y)))
x,y = np.meshgrid(x,y)     
fig = plt.figure(figsize=(12,8))

#Display the plot
ax = fig.add_subplot(111, projection='3d')
p = ax.plot_surface(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('PDF')
ax.set_title('2_D Gaussian')
plt.show()
```
